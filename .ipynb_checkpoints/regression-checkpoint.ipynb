# Machine Learning (Regression)

## Library Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import root_mean_squared_error

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor

from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

## Data Import

```python
df = pd.read_csv('./data/cleaned_taxi.csv')
```

```python
df.head()
```

## Machine Learning

```python
def model_factory(model, nom_columns):
    """
    Creator: Daniel
    inputs: 
        model: The machine learning model object to use in the pipeline
        columns: the list of columns to use for one hot encoding
    outputs:
        returns the model, ready to be fitted and used. 
    """
    preprocessor = ColumnTransformer(
        transformers=[
            ('nom', OneHotEncoder(handle_unknown='ignore'), nom_columns)
        ],
        remainder='passthrough'
    )

    pipe_model = Pipeline(
        steps=[
            ('preprocess', preprocessor),
            ('model', model)
        ]
    )

    return pipe_model
```

```python
columns = ['distance', 'pickup_borough', 'dropoff_borough', 'trip_hour', 'trip_day_num'] # columns used for modeling
nom_cols = ['pickup_borough', 'dropoff_borough'] # nominal columns for one hot encoding
```

```python
X = df[columns] # independant variables
y = df['fare'] # dependant variables
```

```python
scores = [] # keep track of the r2 and rmse scores of linear regression and random forest. 
for size in np.arange(0.2, 0.4, 0.01):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=size, random_state=42)
    # splits the data into training and testing data
    lr = model_factory(LinearRegression(), nom_cols)
    # linear regression model pipeline
    rf = model_factory(RandomForestRegressor(random_state=42), nom_cols)
    # random forest regression model pipeline
    lr.fit(X_train, y_train) # fits the linear regression model
    rf.fit(X_train, y_train) # fits the random forest regression model
    lr_score = lr.score(X_test, y_test) # gets the r2 score of the lr model
    rf_score = rf.score(X_test, y_test) # gets the r2 score of the rf model

    lr_preds = lr.predict(X_test) # makes predictions of the lr model
    lr_rmse = root_mean_squared_error(y_test, lr_preds) # gets the lr rmse score
    rf_preds = rf.predict(X_test) # makes predictions of the rf model
    rf_rmse = root_mean_squared_error(y_test, lr_preds) # gets the rf rmse score

    baseline = np.full_like(y_test, y_test.mean()) # baseline scores
    baseline_rmse = root_mean_squared_error(y_test, baseline) # baseline rmse score
    
    scores.append({'lr_r2': lr_score, 
                   'rf_r2': rf_score, 
                   'lr_rmse': lr_rmse,
                   'rf_rmse': rf_rmse,
                   'baseline_rmse': baseline_rmse,
                   'test_size': size}) # add scores to scores list. 
```

```python
scores_df = pd.DataFrame(scores) # creates scores dataframe from scores list
```

```python
scores_df.sort_values('lr_r2', ascending=False)
```

```python
scores_df.sort_values('rf_r2', ascending=False)
```

```python
sns.lineplot(data=scores_df, y='lr_r2', x='test_size', marker='o', label='Linear Regression')
sns.lineplot(data=scores_df, y='rf_r2', x='test_size', marker='o', label='Random Forest')
plt.legend();
plt.savefig('./images/lr_rf_r2.png')
# shows line plot of r2 score comparison from lr to rf
```

### KNN 

```python
knn_scores = []
for size in np.arange(0.2, 0.4, 0.01):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=size, random_state=42)
    for neighbors in np.arange(3, 31, 2):
        knn = model_factory(KNeighborsRegressor(n_neighbors=neighbors), nom_cols)
        knn.fit(X_train, y_train)
        knn_r2 = knn.score(X_test, y_test)
        knn_preds = knn.predict(X_test)
        knn_rmse = root_mean_squared_error(y_test, knn_preds)
        baseline = np.full_like(y_test, y_test.mean())
        baseline_rmse = root_mean_squared_error(y_test, baseline)
        knn_scores.append({
            'neighbors': neighbors,
            'r2': knn_r2,
            'rmse': knn_rmse,
            'baseline': baseline_rmse,
            'test_size': size
        })
```

```python
knn_df = pd.DataFrame(knn_scores)
```

```python
knn_df.sort_values(by='r2', ascending=False)
```

```python
fig = plt.figure(figsize=(16, 16))
ax = fig.add_subplot(111, projection='3d')

ax.plot(knn_df['test_size'], knn_df['neighbors'], knn_df['r2'], marker='o')

ax.set_xlabel('Test Size')
ax.set_ylabel('Neighbors')
ax.set_zlabel('R2 Score')
ax.set_title("R2 Scores of KNN Model");
# ax.view_init(elev=0, azim=45)
plt.savefig('./images/knn_r2.png')
```

The best performing model was KNN, with a neighbors size of 17, and a test size of 0.39
